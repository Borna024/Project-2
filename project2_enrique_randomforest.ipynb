```python
concat_df = concat_df.drop(columns=['ticker', 'date'])
concat_df.head(14)
```

```python
#concat_df = concat_df.astype("float64").dtypes 
#print(df.dtypes) 
#concat_df.astype('float64')
#print (concat_df.dtypes)
#concat_df.dtypes
import numpy as np
#new_df = pd.DataFrame()
for i in concat_df.columns:
    new_df[i] = pd.to_numeric(concat_df[i],downcast='float')

#new_df.dtypes
new_df = new_df.fillna(new_df.mean())
```

```python
#concat_binary_encoded = pd.get_dummies(concat_df, columns=["target"])
#concat_binary_encoded.tail()
```

```python
# Define features set
X = new_df.copy()
X.drop("target", axis=1, inplace=True)
X.head()
```

```python
# Define target vector
y = new_df["target"].values.reshape(-1, 1)
y[:5]
```

```python
# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Create the StandardScaler instance
scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
X_scaler = scaler.fit(X_train)
```

```python
# Scale the training data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Fitting the Random Forest Model

Once data is scaled, create a random forest instance and train it with the training data (`X_train_scaled` and `y_train`), define `n_estimators=500` and `random_state=78`.

```python
# Create the random forest classifier instance
rf_model = RandomForestClassifier(n_estimators=500, random_state=78)
```

```python
# Making predictions using the testing data
predictions = rf_model.predict(X_test_scaled)
```

```python
# Fit the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

## Model Evaluation

Evaluate model's results, by using `sklearn` to calculate the confusion matrix, the accuracy score and to generate the classification report.

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
